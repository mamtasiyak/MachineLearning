```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
dir(diabetes)
```

```python
#print(diabetes.DESCR)
#print(diabetes.data)
#diabetes.data_filename
print(diabetes.feature_names)
#print(diabetes.target)
#print(diabetes.target_filename)
```

```python
feature_attributes=diabetes.feature_names
feature_attributes
```

```python
features=diabetes.data
print(type(features))
print(features.shape)
label=diabetes.target
#print(label)
```

```python
age=features[:,0]
sex=features[:,1]
bmi=features[:,2]
bp=features[:,3]
s1=features[:,4]
s2=features[:,5]
s3=features[:,6]
s4=features[:,7]
s5=features[:,8]
s6=features[:,9]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.pie(age,sex)
plt.show()
```

```python
plt.scatter(bmi,bp,label='bmp')
plt.scatter(s1,s2,label='s12')
plt.scatter(s2,s3,label='s23')
plt.scatter(s3,s4,label='s34')
plt.scatter(s4,s5,label='s45')
plt.scatter(s5,s6,label='s56')
plt.scatter(age,sex,label='ag')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_fea,test_fea,train_label,test_label=train_test_split(features,label,train_size=.85)
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
trained=clf.fit(train_fea,train_label)
```

```python
predicted=trained.predict(test_fea)
#print(predicted)
#print(test_label)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_label,predicted)
```
