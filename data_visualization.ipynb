```python
import matplotlib.pyplot as plt
```

```python
# list
x1=[2,5,3,4]
y1=[5,3,5,8]

x2=[4,6,3,2]
y2=[7,4,5,3]
```

```python
plt.xlabel("Time",c='red') # x axis label
plt.ylabel("Distance",c='blue') # y axix label
plt.grid(c='green')  # to put graph lines 
plt.plot(x1,y1,label="cars",c='magenta')
plt.plot(x2,y2,label="bikes",c='yellow')
plt.legend()  # to show the lable of plot
```

# Bar plots

```python
plt.xlabel('time',c='red')
plt.ylabel('price',c='blue')
plt.bar(x2,y2,label="Microsoft")
plt.bar(x1,y1,label='Samsung')
plt.plot(x1,y1,label="Amazon",c='cyan')
plt.grid(c="green")
plt.legend()
```

```python
players=["Yuvraj","Dhoni","Sachin"]
Runs=[234,345,456]
```

```python
plt.bar(players,Runs,label="cricket")
plt.xlabel("players",c="cyan")
plt.ylabel("runs",c="magenta")
plt.grid(c="Red")
plt.legend()
```

# Scatter/dot plots


```python
plt.scatter(x1,y1,c="magenta",marker='*',s=100)
plt.grid(c='y')
plt.scatter(x1,y2,c='c',marker='^',s=120)
plt.plot(x1,y1)
plt.plot(x2,y2)
```
